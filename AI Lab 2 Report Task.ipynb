Find Total Impedance 

i. Add the three impedances. 

ii. Print real part (resistive) and imaginary part (reactive). 

iii. Print magnitude and phase angle (use abs() and cmath.phase()). 

```python
import math, cmath

r = 10
l = 7
c = 5
f = 50

w = 2 * math.pi * f

z = complex(r, w*l - 1/(w*c))

print("Total impedance:", z)
print("Real part:", z.real)
print("Imaginary part:", z.imag)
print("Magnitude:", abs(z))
print("Phase angle:", cmath.phase(z))
```
