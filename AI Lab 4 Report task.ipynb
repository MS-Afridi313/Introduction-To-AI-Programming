In circuit design, you often calculate series resistance.

•	Write a function series() that takes a list or tuple of resistances.

•	Return the total resistance.

•	Add docstrings and annotations.

•	Print the function’s __annotations__.


```python
def series(resistances: list | tuple) -> float:
    """Calculate total resistance of resistors connected in series."""
    return sum(resistances)

print(series.__annotations__)
```
