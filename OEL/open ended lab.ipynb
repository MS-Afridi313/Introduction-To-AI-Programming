```python
import csv
import os


# CEP TASK 1: LOAD DATA FILE


building_file = r"E:\Introduction to Ai programming\CEP\citylearn_challenge_2022_phase_all\Building_2.csv"

if not os.path.exists(building_file):
    print("Error: Building file not found!")

else:
   
    # CEP TASK 2: LOAD CONSUMPTION DATA
 

    consumption = []

    with open(building_file, 'r', newline='', encoding='latin-1') as f:
        reader = csv.DictReader(f)
        for row in reader:
            consumption.append(
                float(row["Equipment Electric Power [kWh]"])
            )

    # CEP TASK 3: ANNUAL CONSUMPTION
  

    annual_consumption = sum(consumption)

  
    # CEP TASK 4: DAILY CONSUMPTION
    # (24 hours = 1 day)


    daily_consumption = []

    for i in range(0, len(consumption), 24):
        daily_consumption.append(
            sum(consumption[i:i + 24])
        )

    # CEP TASK 5: MONTHLY CONSUMPTION


    days_in_month = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
    monthly_consumption = []

    index = 0
    for days in days_in_month:
        monthly_consumption.append(
            sum(daily_consumption[index:index + days])
        )
        index += days

    
    # CEP TASK 6: SEASONAL CONSUMPTION
  

    winter = monthly_consumption[11] + monthly_consumption[0] + monthly_consumption[1]
    spring = monthly_consumption[2] + monthly_consumption[3] + monthly_consumption[4]
    summer = monthly_consumption[5] + monthly_consumption[6] + monthly_consumption[7]
    autumn = monthly_consumption[8] + monthly_consumption[9] + monthly_consumption[10]

    
    # CEP TASK 7: WEEKDAY VS WEEKEND
   

    weekday_consumption = []
    weekend_consumption = []

    for i in range(len(daily_consumption)):
        if (i + 1) % 7 == 6 or (i + 1) % 7 == 0:
            weekend_consumption.append(daily_consumption[i])
        else:
            weekday_consumption.append(daily_consumption[i])

 
    # FINAL OUTPUT


    print("Annual Consumption (kWh):", annual_consumption)
    print("Monthly Consumption (kWh):", monthly_consumption)

    print("Winter Consumption (kWh):", winter)
    print("Spring Consumption (kWh):", spring)
    print("Summer Consumption (kWh):", summer)
    print("Autumn Consumption (kWh):", autumn)

    print("Average Weekday Consumption (kWh):",
          sum(weekday_consumption) / len(weekday_consumption))

    print("Average Weekend Consumption (kWh):",
          sum(weekend_consumption) / len(weekend_consumption))
```
