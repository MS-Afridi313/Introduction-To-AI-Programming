```python
import csv
import os

# CEP TASK 1: LOAD DATA FILES


building_file = r"E:\Introduction to Ai programming\CEP\citylearn_challenge_2022_phase_all\Building_2.csv"
price_file = r"E:\Introduction to Ai programming\CEP\citylearn_challenge_2022_phase_all\pricing.csv"

if not os.path.exists(building_file):
    print("Building file not found!")
elif not os.path.exists(price_file):
    print("Price file not found!")
else:

    # CEP TASK 2: ELECTRICITY CONSUMPTION

    non_shiftable_load = []
    solar_generation = []

    with open(building_file, 'r', newline='') as f:
        reader = csv.DictReader(f)
        for row in reader:
            non_shiftable_load.append(
                float(row["Equipment Electric Power [kWh]"])
            )
            solar_generation.append(
                float(row["Solar Generation [W/kW]"])
            )

    total_consumption = sum(non_shiftable_load)


    # CEP TASK 3: PV GENERATION CONVERSION


    net_solar_generation_building_2 = []

    for value in solar_generation:
        pv = (4 * value) / 1000
        net_solar_generation_building_2.append(pv)

    # CEP TASK 4: BILL WITHOUT PV
    

    price = []

    with open(price_file, 'r', newline='') as f:
        reader = csv.DictReader(f)
        for row in reader:
            price.append(float(row["Electricity Pricing [$]"]))

    bill_without_pv = []

    for i in range(len(non_shiftable_load)):
        bill_without_pv.append(
            non_shiftable_load[i] * price[i]
        )

    total_bill_without_pv = sum(bill_without_pv)

    # CEP TASK 5: NET CONSUMPTION & SPLITTING


    net_consumption = []

    for i in range(len(non_shiftable_load)):
        net_consumption.append(
            non_shiftable_load[i] - net_solar_generation_building_2[i]
        )

    net_electricity_consumption_with_pv = []
    wasted_energy = []

    for value in net_consumption:
        if value >= 0:
            net_electricity_consumption_with_pv.append(value)
        else:
            wasted_energy.append(value)

   
    # FINAL OUTPUT


    print("Total Electricity Consumption (kWh):", total_consumption)
    print("Total Bill Without PV ($):", total_bill_without_pv)
    print("Net Electricity Consumption With PV (kWh):",
          sum(net_electricity_consumption_with_pv))
    print("Total Wasted Energy (kWh):", sum(wasted_energy))
```
